## 2. Análise Univariada: Entendendo Variáveis Individuais

Vamos começar analisando as variáveis uma de cada vez para entender suas distribuições.

### A. Histograma: Distribuição de uma Variável Numérica

**Conceito:** O histograma é ideal para visualizar a distribuição de uma única variável numérica. Ele agrupa os números em intervalos (bins) e mostra a frequência de observações em cada intervalo.

**Sua Tarefa:** O gerente quer saber o perfil do valor das contas. A maioria das contas é de valor baixo, médio ou alto? Use um histograma para visualizar a distribuição da coluna `total_bill`.

**Sua Implementação:**

```python
# SEU CÓDIGO AQUI
# Crie um histograma para a coluna 'total_bill'.
# Dica: Use sns.histplot() e plt.show(). Não se esqueça de adicionar um título com plt.title().
plt.figure(figsize=(8, 5))
sns.histplot(data=df, x='total_bill', bins=20, kde=True, color='skyblue')
plt.title('Distribuição do Valor Total da Conta (Total Bill)', fontsize=14)
plt.xlabel('Valor total da conta (US$)')
plt.ylabel('Frequência')
plt.show()
```

```python
plt.figure(figsize=(8, 5))
sns.histplot(data=df, x='total_bill', bins=20, kde=False, color='steelblue')
# média e mediana
media = df['total_bill'].mean()
mediana = df['total_bill'].median()
plt.axvline(media, color='red', linestyle='-', linewidth=2, label=f'Média ({media:.1f})')
plt.axvline(mediana, color='black', linestyle='--', linewidth=2, label=f'Mediana ({mediana:.1f})')
# título e rótulos
plt.title('Histograma - Total da Conta (US$)', fontsize=14)
plt.xlabel('Valor total da conta (US$)')
plt.ylabel('Frequência')
plt.legend()
plt.tight_layout()
plt.show()
```

### B. Box Plot: Comparando Distribuições

**Conceito:** O box plot é excelente para resumir a distribuição de uma variável numérica e é especialmente poderoso para comparar distribuições entre diferentes categorias. Ele exibe a mediana, os quartis e os outliers de forma compacta.

**Sua Tarefa:** O gerente suspeita que os clientes gastam mais nos fins de semana. Compare a distribuição do valor total da conta (`total_bill`) para cada dia da semana (`day`).

**Sua Implementação:**

## 3. Análise Bivariada: Explorando Relações

Agora, vamos investigar a relação entre pares de variáveis.

### C. Gráfico de Dispersão (Scatter Plot): Relação Entre Duas Variáveis Numéricas

**Conceito:** O gráfico de dispersão é a melhor ferramenta para visualizar a relação entre duas variáveis numéricas. Cada ponto no gráfico representa uma observação.

**Sua Tarefa:** Uma pergunta fundamental para o negócio é: "Clientes que gastam mais na conta também dão gorjetas maiores?". Crie um gráfico de dispersão para visualizar a relação entre `total_bill` e `tip`.

**Sua Implementação:**

```python
# SEU CÓDIGO AQUI
# Crie um gráfico de dispersão com 'total_bill' no eixo X e 'tip' no eixo Y.
# Dica: Use sns.scatterplot(x='total_bill', y='tip', data=df).
# Criando um gráfico de dispersão entre total_bill e tip
plt.figure(figsize=(8, 5))
sns.scatterplot(x='total_bill', y='tip', data=df, color='steelblue')
plt.title('Relação Valor Total da Conta e Valor da Gorjeta', fontsize=14)
plt.xlabel('Valor Total da Conta (US$)')
plt.ylabel('Valor da Gorjeta (US$)')
plt.show()
```

```python
# covariância e correlação de Pearson entre total_bill e tip
cov = df["total_bill"].cov(df["tip"])
r = df["total_bill"].corr(df["tip"])

print(f"Covariância: {cov:.2f}")
print(f"Correlação de Pearson (r): {r:.2f}")

# interpretação orientativa
if r >= 0.9:
    intensidade = "muito forte"
elif r >= 0.7:
    intensidade = "forte"
elif r >= 0.3:
    intensidade = "moderada"
elif r > 0:
    intensidade = "fraca"
elif r == 0:
    intensidade = "nula"
elif r > -0.3:
    intensidade = "fraca (negativa)"
elif r > -0.7:
    intensidade = "moderada (negativa)"
elif r > -0.9:
    intensidade = "forte (negativa)"
else:
    intensidade = "muito forte (negativa)"

print(f"Interpretação: relação {intensidade}.")
```

**Sua Análise e Conclusão:**
*Escreva sua análise aqui...*
* Você observa alguma relação entre o valor da conta e o valor da gorjeta?
Descreva o padrão.

Sim existe uma relação entre o valor da conta e o valor da gorjeta, a medida que o valor total da conta aumenta, o valor da gorjeta também tende a aumentar.

* A relação parece ser forte ou fraca? É linear?

Acredito que seja modereado/forte mais não é algo linear

* O que isso significa para o restaurante?

Bom, isso mostra que estimular o consumo pode aumentar as gorjetas mas um bom atendimento ainda é essencial.

### D. Gráfico de Barras: Comparando Médias Entre Categorias

**Conceito:** O gráfico de barras é usado para comparar uma métrica numérica (como média ou soma) entre diferentes grupos categóricos.

**Sua Tarefa:** O gerente quer saber se há diferença no valor médio das contas entre almoços (`Lunch`) e jantares (`Dinner`). Use um gráfico de barras para comparar o `total_bill` médio para cada período (`time`).

**Sua Implementação:**

```python
# SEU CÓDIGO AQUI
# Crie um gráfico de barras que mostre o 'total_bill' médio por 'time'.
# Dica: Use sns.barplot(x='time', y='total_bill', data=df). O Seaborn calcula a média por padrão.
plt.figure(figsize=(7, 5))
sns.barplot(x='time', y='total_bill', data=df, palette='pastel')
plt.title('Valor Médio da Conta por Período (Almoço vs Jantar)', fontsize=14)
plt.xlabel('Período da Refeição')
plt.ylabel('Valor Médio da Conta (US$)')
plt.show()
```
